```python
import zipfile

# Path to the zip file
zip_file_path = output_path

# Destination directory to extract the contents
extract_dir = 'result/Exp/PhotoSketch/Checkpoints/pretrained/'

# Open the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    # Extract all the contents to the specified directory
    zip_ref.extractall(extract_dir)
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import os

# Path to the folders containing the images
folder1_path = 'examples'
folder2_path = 'result/Exp/PhotoSketch/Results'

# Get the list of image files in folder 1
folder1_files = os.listdir(folder1_path)

# Iterate over the image files in folder 1
for filename in folder1_files:
    # Construct the file paths for the images in folder 1 and folder 2
    folder1_file_path = os.path.join(folder1_path, filename)
    folder2_file_path = os.path.join(folder2_path, filename.split('.')[0] + '.png')

    # Open the images
    img1 = Image.open(folder1_file_path)
    img2 = Image.open(folder2_file_path)

    # Create a figure with two subplots
    fig, axs = plt.subplots(1, 2)

    # Display the images side by side
    axs[0].imshow(img1)
    axs[0].axis('off')
    axs[0].set_title('Folder 1')

    axs[1].imshow(img2)
    axs[1].axis('off')
    axs[1].set_title('Folder 2')

    # Show the figure
    plt.show()
```
